```python
import math
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.python.framework import ops
import os
from PIL import Image
```

```python
DIR = 'Dataset/Guts'
path = os.path.join(DIR)

X_ = []
Y_ = []

for img in os.listdir(path):
    image = Image.open(os.path.join(path, img))
    data = np.array(image)
    X_.append(data)
    Y_.append(0)

DIR = "Dataset\Griffith"
path = os.path.join(DIR)


for img in os.listdir(path):
    image = Image.open(os.path.join(path, img))
    data = np.array(image)
    X_.append(data)
    Y_.append(1)
    
m = 200
```

```python
X = np.array(X_)
Y = np.array(Y_)


permutation = list(np.random.permutation(m))
sX = X[permutation]
sY = Y[permutation].reshape((1,m))
```
